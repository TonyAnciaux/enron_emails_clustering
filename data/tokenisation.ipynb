### Creating dictionary that contain the tokens for each email 

```python
def token_of_email(email):
    tokens = word_tokenize(email)
    return [w for w in tokens if not w.lower() in stop_words and w.isalpha()]
```

```python
df2 = df.iloc[:100,:]
```

```python
type(df2)
```

```python
df2.shape
```

```python
df2["token"] = df2['message'].map(token_of_email)
```

```python
df2.shape
```

```python
df2["counter"] = df2["token"].map(lambda x: Counter(x))
```

## LDA (Latent Dirichlet Allocation)

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.datasets import make_multilabel_classification
```


Methods

fit(X[, y])
Learn model for the data X with variational Bayes method.
fit_transform(X[, y])
Fit to data, then transform it.
get_params([deep])
Get parameters for this estimator.
partial_fit(X[, y])
Online VB with Mini-Batch update.
perplexity(X[, sub_sampling])
Calculate approximate perplexity for data X.
score(X[, y])
Calculate approximate log-likelihood as score.
set_params(**params)
Set the parameters of this estimator.
transform(X)
Transform data X according to the fitted model.
